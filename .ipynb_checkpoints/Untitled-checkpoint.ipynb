```python
file_name = [];
file_size = [];
completed_labels = []



for i in glob.glob('All_Data/*.xlsx'):
    try:
        file = pd.read_excel(i)
        if file.shape[0] != 0:
            file_name.append(os.path.basename(i))
            print(i)
            file_size.append(file.shape[0])
            count += file.shape[0]
            
            
            
    except:
        continue
```
